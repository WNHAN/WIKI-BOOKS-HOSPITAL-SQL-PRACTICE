```python
## Create table Department

import mysql.connector

#establishing the connection
mydb = mysql.connector.connect(
host = "localhost",
user = 'root',
passwd = "",
database ="Hospital")

#Creating a cursor object using the cursor() method
mycursor = mydb.cursor()

#Dropping Department talbe if already exists.
mycursor.execute("DROP TABLE IF EXISTS Department")

#Creating table as per requirement
sql = '''CREATE TABLE Department(
DepartmentID INTEGER NOT NULL,
Name Varchar(30) NOT NULL,
Head INTEGER NOT NULL,
CONSTRAINT pk_Department PRIMARY KEY(DepartmentID),
CONSTRAINT fk_Department_Physician_EmployeeID FOREIGN KEY(Head) REFERENCES Physician(EmployeeID))'''

mycursor.execute(sql)

#closing the connection
mydb.close()
```

```python
## Question 1: Obtain the names of all physicians that have performed a medical procedure they have never been certified to perform.

import mysql.connector

#establishing the connection
mydb = mysql.connector.connect(
   user='root', password='1234#,', host='localhost', database='Hospital')

#Creating a cursor object using the cursor() method
mycursor = mydb.cursor()

#Retrieving single row
sql = '''SELECT Name 
  FROM Physician 
 WHERE EmployeeID IN 
       (
         SELECT Physician FROM Undergoes U WHERE NOT EXISTS 
            (
               SELECT * FROM Trained_In 
                WHERE Treatment = Procedures 
                  AND Physician = U.Physician
            )
       )'''

#Executing the query
mycursor.execute(sql)

#Fetching 1st row from the table
question1 = mycursor.fetchall();
print(question1)
```

```python
## Question 2: Same as the query in Question 1, but include the following information in the results: Physician name, name of procedure, 
## date when the procedure was carried out, name of the patient the procedure was carried out on. 

import mysql.connector

#establishing the connection
mydb = mysql.connector.connect(
    user='root',
    password='1234#,', 
    host='localhost', 
    database='Hospital')

#Creating a cursor object using the cursor() method
mycursor = mydb.cursor()

#Retrieving single row
sql = '''SELECT Physician.Name,
                Procedures.Name,
                Undergoes.Date,
                Patient.Name
        FROM Physician, Procedures, Undergoes, Patient
        WHERE undergoes.patient = patient.ssn
        AND  UNdergoes.physician = physician.employeeID
        AND  undergoes.procedures = procedures.code
        AND  NOT EXISTS
(SELECT * FROM Trained_In
          WHERE trained_in.physician = undergoes.physician
          AND Trained_in.treatment = undergoes.procedures)
'''

#Executing the query
mycursor.execute(sql)

#Fetching 1st row from the table
question2 = mycursor.fetchall();
print(question2)
```

```python
## Question 3: Obtain the names of all physicians that have performed a medical procedure that they are certified to perform, but such that 
## the procedure was done at a date (Undergoes.Date) after the physician's certification expired (Trained_In.CertificationExpires). 

import mysql.connector

#establishing the connection
mydb = mysql.connector.connect(
    user='root',
    password='', 
    host='localhost', 
    database='Hospital')

#Creating a cursor object using the cursor() method
mycursor = mydb.cursor()

#Retrieving single row
sql =''' SELECT Name     
         FROM Physician
         WHERE EmployeeID in
              (SELECT Physician FROM Undergoes
                 WHERE Undergoes.date > 
                   (SELECT CertificationExpires FROM Trained_in
                    WHERE Undergoes.Procedures = Trained_in.Treatment AND
                    Undergoes.Physician = Trained_in.Physician))
         '''

#Executing the query
mycursor.execute(sql)

#Fetching 1st row from the table
question3 = mycursor.fetchall();
print(question3)
```
